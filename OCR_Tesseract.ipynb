```python
### Helper Functions
def clean_spaces(input_string):
    space_to_empty = [x.strip() for x in input_string.split('\n')]
    space_clean_list = [x for x in space_to_empty if x]

    return space_clean_list
```

```python
#Setting up libraries
from pyimagesearch.transform import four_point_transform
from skimage.filters import threshold_local
import numpy as np
import argparse
import cv2
import imutils
import pandas as pd
from PIL import Image
import os
import shutil

#Reading Bill
Invoice_Name= "bill (1).jpg"
image = cv2.imread("Input_format1/"+ Invoice_Name)
orig = image.copy()
```

```python
#Bill processing
#Converting bills to gray scale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray = cv2.GaussianBlur(gray, (5, 5), 0)
edged = cv2.Canny(gray, 75, 200)


#Finding contors
cnts = cv2.findContours(edged.copy(), cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
cnts = imutils.grab_contours(cnts)
cnts = sorted(cnts, key = cv2.contourArea, reverse = True)[:5]
 
for c in cnts:
    # approximate the contour
    peri = cv2.arcLength(c, True)
    approx = cv2.approxPolyDP(c, 0.02 * peri, True)
    if len(approx) == 4:
        screenCnt = approx
        break


#Adjustment of bill according to contors
warped = four_point_transform(orig, screenCnt.reshape(4, 2))
warped = cv2.cvtColor(warped, cv2.COLOR_BGR2GRAY)
T = threshold_local(warped, 11, offset = 10, method = "gaussian")
warped = (warped > T).astype("uint8") * 255
```

```python
# show the original and scanned images
#cv2.imwrite('final1.png', imutils.resize(orig, height = 650))
cv2.imwrite('im1.jpg', imutils.resize(warped, height = 650))
```

```python
img = Image.open(Image_Path)
result = pytesseract.image_to_string(img)
l = result.split('\n')
```

```python
l
```

```python
import pandas as pd
xl = pd.ExcelFile("seller.xlsx")
xl.sheet_names
df_s = xl.parse("Sheet1")
#df_s
```

```python
from difflib import get_close_matches 

seller = []
for j in range(len(df_s)): 
    Seller_name = get_close_matches(df_s['Seller Name'][j],l)
    seller = seller + Seller_name
```

```python
start = ['description', 'descr1ption', 'units']
end = ['total','totals','grand total','sgst', 'IGST']
```

```python
index_DESC = [idx for idx, s in enumerate(l) if any(word in s.lower() for word in start)][0]
index_Total = [idx for idx, s in enumerate(l) if any(word in s.lower() for word in end)][-1]

desc = l[index_DESC+1:index_Total+1]
space_to_empty = [x.strip() for x in desc]
space_clean_list = [x for x in space_to_empty if x]
```

```python
#if len(desc)>=1:
space_clean_list = [item for item in space_clean_list if item[-1].isdigit()] 

amt = []
for word in space_clean_list:
    word = word.split(' ') 
    amt_one = word[-1]
    amt = amt + [amt_one]
```

```python
lineItems =[]
for word in space_clean_list:
    item = word.rsplit(' ', 1)[0]
    lineItems = lineItems + [item]
```

# Creating dataframe

```python
import pandas as pd
import numpy as np
frame =pd.DataFrame()

frame['lineItems']=lineItems
frame['amount'] =amt


frame['amount'].replace('', np.nan, inplace=True)
frame.dropna(subset=['amount'], inplace=True)

frame['lineItems'] = frame['lineItems'].str.findall('([A-Z]\w{0,})')


frame = frame.reset_index(drop=True)
```

```python
for i in range(len(frame['lineItems'])):
    frame['lineItems'][i] = ' '.join(frame['lineItems'][i])
```
